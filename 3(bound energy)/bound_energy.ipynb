```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
r = []
E = []
E_bond = -14.8
for file in os.listdir('bound_energy_sigma/OUTCAR'):
    print(file, file.find('OUTCAR'))
    if file.find('_')!=-1 and file.find('bond')==-1:
        r0 = float(file[file.find('_')+1:].replace('p', '.'))
        #print(r0)
        r.append(r0)
        with open(os.path.join('bound_energy_sigma/OUTCAR', file)) as f:
            line=''
            for line in f:
                if 'energy(sigma->0) =      ' in line:
                    line_E = line
            if 'energy(sigma->0) =      ' in line_E:
                E0 = float(line_E[line_E.index('energy(sigma->0) =      ')+len('energy(sigma->0) =      '):-2])
            E.append(E0)
        #print(E0)
r = np.array(r)
E = np.array(E)
print(r, E)
```

```python
from scipy.interpolate import interp1d
interp = interp1d(r,(E-E_bond)/E_bond, kind='cubic')
x = np.linspace(r[0], r[-1], 500)

%matplotlib notebook
plt.plot(x, interp(x), label='interpolation')
plt.plot(r,(E-E_bond)/E_bond, '.', label='data')
plt.plot(1.12, 0, '.')
plt.title(r'CO')
plt.xlabel('r, A')
plt.ylabel(r'$\frac{E-E_{bond}}{E_{bond}}$')
plt.grid()
plt.legend()
```

```python
from scipy.interpolate import interp1d
interp = interp1d(r, -E/E_bond, kind='cubic')
x = np.linspace(r[0], r[-1], 500)

%matplotlib notebook
plt.plot(x, interp(x), label='interpolation')
plt.plot(r, -E/E_bond, '.')
plt.title(r'CO')
plt.xlabel('r, A')
plt.ylabel(r'$-\frac{E}{E_{bond}}$')
plt.grid()
```
