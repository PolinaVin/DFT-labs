```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = [3.4, 3.2, 3, 2.9, 2.866, 2.7, 2.5, 2.3, 2.8, 2.82, 2.78, 2.76, 2.84]
E = [-14.19, -15.36, -16.2, -16.43, -16.48, -16.19, -13.77, -7.15, -16.501, -16.51, -16.476, -16.435, -16.5046]
r = np.array(r)
E = np.array(E)
```

```python
from scipy.interpolate import interp1d
interp = interp1d(r, E, kind='cubic')
x = np.linspace(np.min(r), np.max(r), 500)
y = interp(x)

x_min = x[y==np.min(y)]
y_min = np.min(y)

%matplotlib notebook
plt.plot(x, y, color='C2', zorder=-1, label='interporation')
plt.scatter(r, E, label='data')
plt.scatter(x_min, y_min, label=r'r = '+str(np.round(x_min, 2)[0])+r' $\AA$')
plt.xlabel(r'Lattice constant, $\AA$')
plt.ylabel(r'Energy, eV')
plt.legend()
plt.grid()
```
